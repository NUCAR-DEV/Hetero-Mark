```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('out.data', header=None);data
```

```python
lines = []
def iterate_row(data):
    line = []
    for i in range(0, len(data)/2):
        point = (data[i*2], data[i*2 + 1])
        line.append(point)
    lines.append(line)
data.apply(iterate_row, axis=1)
lines
```

```python
line_collections = mpl.collections.LineCollection(lines)
fig, ax = plt.subplots(figsize=(10, 10))
ax.add_collection(line_collections)
plt.xlim([0, 1000])
plt.ylim([0, 1000])
plt.savefig('out_bundle.eps')
```
