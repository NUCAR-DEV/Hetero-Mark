```python
lines_in = []
def iterate_row(data):
    line = []
    for i in range(0, len(data)/2):
        point = (data[i*2], data[i*2 + 1])
        line.append(point)
    lines_in.append(line)
data_in.apply(iterate_row, axis=1);
```
